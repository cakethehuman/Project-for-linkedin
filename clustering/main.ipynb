```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\wilsen\OneDrive\Desktop\PROJECT PORTO\clustering\segmentation data (1).csv")
```

```python
df = df.drop(['ID'], axis=1)
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()

x = df.columns
x_scaled = scale.fit_transform(x)
```

```python
model = KMeans(n_clusters= 3, random_state= 42)
df['Clustering'] = model.fit_predict(df[df.columns])
```

```python
centers = model.cluster_centers_
centers
```

```python
df.columns
```

```python
plt.scatter(df['Age'], df['Income'], c=df['Clustering'], cmap='viridis', s=80)
plt.scatter(centers[:, 2], centers[:, 4], c='y', marker='X', s=200)
plt.title("Customer Segmentation (Unsupervised Learning)")
plt.show()
```
